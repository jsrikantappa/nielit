```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
sns.distplot(df['tip']);
```

```python
sns.distplot(df['tip'],kde=False);
```

```python
sns.distplot(df['tip'],hist=False);
```

```python
sns.jointplot(x='tip',y='total_bill',data=df);
```

```python
sns.pairplot(df,hue='sex');
```

```python
sns.stripplot(x= 'time',y='tip',data=df);
```

```python
sns.boxplot(x= 'time',y='tip',data=df);
```

```python
sns.boxplot(x= 'sex',y='tip',data=df);
```

```python
sns.barplot(x='smoker',y='tip',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)
```
