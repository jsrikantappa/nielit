```python
class Atm:
    
    #Using default balance as 10000 can change this if needed.
    global balance
    balance = 10000 
    
    def __init__(self):
        
        #We will get the input of the Pin from the user
        print("PIN should be exactly 4 digit\n")
        self.pin1 = input("\nPlease input Your PIN: \n")
        self.pin2 = input("\nRe-Enter you PIN: \n")
    
    #When called shows the existing balance
    def saving_b(self,balance):
        print("\n Your existing Savings Balance is \n",balance)
        a=self.options(balance)
        return a
        
    #When called accepts the deposits and updates the balance
    def Deposit(self,balance):
        print("\nYour existing Savings Balance is \n",balance)
        print("\nYou are Deposting funds to your Savings\n")
        a=input("Please enter the amount you want to Deposit\n")
        balance += int(a)
        print("Your Savings Balance is updated to ",balance)
        a=self.options(balance)
        return a
    
    #When called withdraws the amount from savings balance and updates the balance
    def Withdraw(self,balance):
        print("\nYour existing Savings Balance is \n",balance)
        print("\nYou are Withdrawing funds from Savings\n")
        a=input("\nPlease enter the amount you want to withdraw\n")
        balance -= int(a)
        if (balance < 0):
            print("\nBalance cannot be less than $0.00\n")
            balance += int(a)
            a=self.options(balance)
            return a
        print("Your Balance is updated to ",balance)
        a=self.options(balance)
        return a
    
    #When called debits the entered amount from Checking balance and credits it to Saving Balance
    def Transfer(self,balance):
        chk_balance  = 10000
        print("\nYour Available checking balance is \n", chk_balance)
        print("\nYour existing Savings Balance is \n",balance)
        a=input("\nPlease enter the amount you want to transfer from Checking Balance \n")
        chk_balance -= int(a)
        if (chk_balance < 0):
            print("\nChecking Balance cannot be less than $0.00\n")
            chk_balance += int(a)
            a=self.options(balance)
            return a
        balance += int(a)
        print("\nYour Savings Balance is updated to \n",balance)
        print("\nYour Checking Balance is updated to \n",chk_balance)
        a=self.options(balance)
        return a
    
    #Input Values 1-4 in turns calls the appropriate method for the functionality like Deposit or Withdraw etc 
    def options(self,balance):
        
        while True:
            
            opt=input("Enter the options from 1-4 \n\n 1= Check saving balance \n 2= Deposit funds \n 3= Withdraw funds \n 4= Transfer funds \n\n")
            
            if ( opt == '1' ):
                a=self.saving_b(balance)
                return a
                
            if ( opt == '2' ):
                a=self.Deposit(balance)
                return a
                
            if ( opt == '3' ):
                a=self.Withdraw(balance)
                return a
            
            if ( opt == '4' ):
                a=self.Transfer(balance)
                return a
            else:
                print('\nInvalid Choice\n')
                break
    
    #Check for pin validity 
    def pin_check(self):
        pin1 =  self.pin1
        pin2 =  self.pin2
        if((len(pin1)==4 or len(pin2)==4) and (pin1==pin2)):
            print('Pin OK')
            a=self.options(balance)
            return a
        else:
            print('\n\n  !!!!ALERT-INVALID PIN!!!!\n\n !! BOTH PIN SHOULD MATCH !! \n\n             &       \n \n   PIN SHOULD BE 4 DIGITS\n    !!NO CHARS ALLOWED!!')
    
    #Intial point of the class Atm
    def main(self):
        self.pin_check()
        
        

a=Atm()
a.main()
    
    
```
